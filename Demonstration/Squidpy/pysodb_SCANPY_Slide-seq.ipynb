```python
import os
import pandas as pd
import numpy as np
import scanpy as sc
import anndata as ad
import squidpy as sq
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pysodb
```

```python
sc.set_figure_params(vector_friendly=False,format='pdf',transparent=True,dpi=50)
plt.rcParams["figure.figsize"] = (8, 8)

sns.set_style('white')
```

### load data using pysodb

```python
sodb = pysodb.SODB() # Initialization

```

```python
# this dataset is from publication "Dissecting the treatment-naive ecosystem of human melanoma brain metastasis"
```

```python
# link in SODB: https://gene.ai.tencent.com/SpatialOmics/dataset?datasetID=119
```

```python
adata = sodb.load_experiment('Biermann2022Dissecting','GSM6025950_ECM10')
```

### play with Squidpy


#### display clusters

```python
sc.pl.embedding(adata,basis='spatial',color='leiden',s=30)
```

#### Neighborhood enrichment

```python
sq.gr.spatial_neighbors(adata)
sq.gr.nhood_enrichment(adata, cluster_key="leiden")
sq.pl.nhood_enrichment(adata, cluster_key="leiden")
```

#### Interaction matrix and network centralities

```python
sq.gr.interaction_matrix(adata, cluster_key="leiden")
sq.pl.interaction_matrix(adata, cluster_key="leiden")
```

#### centrality score analysis

```python
sq.gr.centrality_scores(
    adata,
    cluster_key="leiden",
)
sq.pl.centrality_scores(adata, cluster_key="leiden", figsize=(20, 5), s=500)
```
