```python
from soview import *
import shutil
import os
import gc
from scipy import sparse
import pandas as pd
```

```python
raw_path = '/aaa/jianhuayao/project2/zhiyuanyuan/PUBDT/0814/Barkley2022Cancer'
```

```python
sample_list = []
for file in os.listdir(f'{raw_path}/raw'):
    if file.endswith('.h5'):
        sample = file.split('_filtered_feature_bc_matrix.h5')[0]
        sample_list.append(sample)
```

```python
sample_list
```

```python
for sample in sample_list:
    sample_path = f'{raw_path}/raw_spatial/{sample}'
    sample_count_source = f'{raw_path}/raw/{sample}_filtered_feature_bc_matrix.h5'
    sample_count_target = f'{sample_path}/filtered_feature_bc_matrix.h5'
    
    sample_scalefactor_source = f'{raw_path}/raw/{sample}_spatial_scalefactors_json.json'
    sample_scalefactor_target = f'{sample_path}/spatial/scalefactors_json.json'
    
    sample_lowimage_source = f'{raw_path}/raw/{sample}_spatial_tissue_lowres_image.png'
    sample_lowimage_target = f'{sample_path}/spatial/tissue_lowres_image.png'

    sample_hiimage_source = f'{raw_path}/raw/{sample}_spatial_tissue_hires_image.png'
    sample_hiimage_target = f'{sample_path}/spatial/tissue_hires_image.png'
    
    sample_position_source = f'{raw_path}/raw/{sample}_spatial_tissue_positions_list.csv'
    sample_position_target = f'{sample_path}/spatial/tissue_positions_list.csv'


    os.makedirs(sample_path, exist_ok=True)
    os.makedirs(f'{sample_path}/spatial', exist_ok=True)


    
    shutil.copyfile(sample_count_source, sample_count_target)
    shutil.copyfile(sample_scalefactor_source, sample_scalefactor_target)    
    shutil.copyfile(sample_hiimage_source, sample_hiimage_target)    
    shutil.copyfile(sample_lowimage_source, sample_lowimage_target)    


    shutil.copyfile(sample_position_source, sample_position_target)    
```

```python
for sample in sample_list:
    adata = sc.read_visium(f'{raw_path}/raw_spatial/{sample}')
    adata.write_h5ad(f'{raw_path}/h5ad/{sample}.h5ad')
```

```python
sc.pl.spatial(adata,color='array_row')
```

```python
sc.pl.embedding(adata,basis='spatial',color='array_row')
```
